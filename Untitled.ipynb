```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 파라미터 설정
n = 50  # 동전 던지기 횟수
p = 0.1 # 앞면이 나올 확률 (불균형)

# 데이터 생성
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(x, pmf)
plt.title('Probability Distribution of an Imbalanced Coin (Low Entropy)')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# PNG 파일로 저장
plt.savefig('imbalanced_coin_dist.png', dpi=300)
print("Graph saved as imbalanced_coin_dist.png")
```
